    b)	(1 point) With a for loop, draw and study the values in each of the remaining W columns. (hint for each w column draw histogram and box plot)

```python

import matplotlib.pyplot as plt

 # axis check the it vertically which is Week and drop the columns contains ALL NAN
df = bb_df.dropna(axis=1, how='all')

 #w_cols is a Series like W1,W2,W3 which contains all rows in the column
w_cols = [c for c in df.columns if c.startswith('w')] 

# using for loop for draw the graphics 
for col in w_cols:
    data = df[col].dropna() # iterate each items within the Series, drop the nan.
    
    # histogram
    plt.figure()
    plt.hist(data, bins=10)
    plt.title(f'Histogram of {col}')
    plt.xlabel('Rank')
    plt.ylabel('Frequency')
    plt.show()
    
    # boxplot
    plt.figure()
    plt.boxplot(data)
    plt.title(f'Box Plot of {col}')
    plt.ylabel('Rank')
    plt.show()
```

10. The maintenance manager at a trucking company wants to build a regression model to forcast the time unyil the first engine overhaul(Time in years) based on four predictor variables for 25 trucks:
    1. anual miles driven (Miles in 1000s)
    2. avrage load weight (Load in tons)
    3. average driving speed (Speed in mph)
    4. oil change interval (Oil in 1000s of miles).

```python
import numpy as np
import pandas as pd
engineDf =  pd.read_excel("https://www.dropbox.com/scl/fi/dcoz9yw3f8yywtzy82f4z/Engine.xlsx?rlkey=n53hfjjsrddywktksj156jra5&dl=1")
engineDf.head()
```

    a) (2 point)find model regression equation (use all four predictor variables)

```python
# your answer
# https://www.statsmodels.org/stable/index.html
import statsmodels.api as sm

# Define predictors and response
X = engineDf[['Miles', 'Load', 'Speed', 'Oil']]  # All four predictors
X = sm.add_constant(X)  # Add intercept term
y = engineDf['Time']  # Target variable

# Fit the regression model
model = sm.OLS(y, X).fit()

# Print summary of the model
print(model.summary())
```

    b) (1point) Predict the time before the first engine overhaul for particular truck driven 60000 miles per year with an average load of 22 tons, an average driving speed of 57 mph, and 18000 miles between oil changes.

```python
# Your answer
intercept = 13.4343
miles_coeff = -0.0896
load_coeff = -0.0732
speed_coeff = 0.0048
oil_coeff = -0.0297

miles_in_thousands = 60
load = 22
speed = 57
oil_in_thousands = 18

time = round(intercept+(miles_coeff*miles_in_thousands)+(load_coeff*load)+(speed_coeff*speed)+(oil_coeff*oil_in_thousands),2)
print("The predicted time before the first engine overhaul is approximately", time, "years.")
```
